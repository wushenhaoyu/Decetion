```python
def load_and_fix_weights(model, weights_path):
    # 加载预训练权重
    state_dict = paddle.load(weights_path)

    # 遍历所有权重并调整维度
    for key, weight in state_dict.items():
        if 'conv' in key or 'proj' in key:  # 针对卷积层的维度调整
            if weight.shape[1] != model.state_dict()[key].shape[1]:  # 判断维度是否不匹配
                # 转置权重，调整维度为 [out_channels, in_channels, kernel_height, kernel_width]
                weight = weight.transpose((1, 0, 2, 3))
                state_dict[key] = weight

        if 'fc' in key:  # 针对全连接层的维度调整
            if weight.shape != model.state_dict()[key].shape:
                # 转置全连接层的权重维度
                weight = weight.T
                state_dict[key] = weight

    # 加载调整后的权重到模型
    model.set_state_dict(state_dict)
    return model
```

```python


keys_dict = {
    "local_net.conv1.weight": "local_net.conv1.weight",
    "local_net.conv1.bias": "local_net.conv1.bias",
    "local_net.mul_blocks.0.gamma_1": "local_net.mul_blocks.0.gamma_1",
    "local_net.mul_blocks.0.gamma_2": "local_net.mul_blocks.0.gamma_2",
    "local_net.mul_blocks.0.pos_embed.weight": "local_net.mul_blocks.0.pos_embed.weight",
    "local_net.mul_blocks.0.pos_embed.bias": "local_net.mul_blocks.0.pos_embed.bias",
    "local_net.mul_blocks.0.norm1.alpha": "local_net.mul_blocks.0.norm1.alpha",
    "local_net.mul_blocks.0.norm1.beta": "local_net.mul_blocks.0.norm1.beta",
    "local_net.mul_blocks.0.norm1.color": "local_net.mul_blocks.0.norm1.color",
    "local_net.mul_blocks.0.conv1.weight": "local_net.mul_blocks.0.conv1.weight",
    "local_net.mul_blocks.0.conv1.bias": "local_net.mul_blocks.0.conv1.bias",
    "local_net.mul_blocks.0.conv2.weight": "local_net.mul_blocks.0.conv2.weight",
    "local_net.mul_blocks.0.conv2.bias": "local_net.mul_blocks.0.conv2.bias",
    "local_net.mul_blocks.0.attn.weight": "local_net.mul_blocks.0.attn.weight",
    "local_net.mul_blocks.0.attn.bias": "local_net.mul_blocks.0.attn.bias",
    "local_net.mul_blocks.0.norm2.alpha": "local_net.mul_blocks.0.norm2.alpha",
    "local_net.mul_blocks.0.norm2.beta": "local_net.mul_blocks.0.norm2.beta",
    "local_net.mul_blocks.0.norm2.color": "local_net.mul_blocks.0.norm2.color",
    "local_net.mul_blocks.0.mlp.fc1.weight": "local_net.mul_blocks.0.mlp.fc1.weight",
    "local_net.mul_blocks.0.mlp.fc1.bias": "local_net.mul_blocks.0.mlp.fc1.bias",
    "local_net.mul_blocks.0.mlp.fc2.weight": "local_net.mul_blocks.0.mlp.fc2.weight",
    "local_net.mul_blocks.0.mlp.fc2.bias": "local_net.mul_blocks.0.mlp.fc2.bias",
    "local_net.mul_blocks.1.gamma_1": "local_net.mul_blocks.1.gamma_1",
    "local_net.mul_blocks.1.gamma_2": "local_net.mul_blocks.1.gamma_2",
    "local_net.mul_blocks.1.pos_embed.weight": "local_net.mul_blocks.1.pos_embed.weight",
    "local_net.mul_blocks.1.pos_embed.bias": "local_net.mul_blocks.1.pos_embed.bias",
    "local_net.mul_blocks.1.norm1.alpha": "local_net.mul_blocks.1.norm1.alpha",
    "local_net.mul_blocks.1.norm1.beta": "local_net.mul_blocks.1.norm1.beta",
    "local_net.mul_blocks.1.norm1.color": "local_net.mul_blocks.1.norm1.color",
    "local_net.mul_blocks.1.conv1.weight": "local_net.mul_blocks.1.conv1.weight",
    "local_net.mul_blocks.1.conv1.bias": "local_net.mul_blocks.1.conv1.bias",
    "local_net.mul_blocks.1.conv2.weight": "local_net.mul_blocks.1.conv2.weight",
    "local_net.mul_blocks.1.conv2.bias": "local_net.mul_blocks.1.conv2.bias",
    "local_net.mul_blocks.1.attn.weight": "local_net.mul_blocks.1.attn.weight",
    "local_net.mul_blocks.1.attn.bias": "local_net.mul_blocks.1.attn.bias",
    "local_net.mul_blocks.1.norm2.alpha": "local_net.mul_blocks.1.norm2.alpha",
    "local_net.mul_blocks.1.norm2.beta": "local_net.mul_blocks.1.norm2.beta",
    "local_net.mul_blocks.1.norm2.color": "local_net.mul_blocks.1.norm2.color",
    "local_net.mul_blocks.1.mlp.fc1.weight": "local_net.mul_blocks.1.mlp.fc1.weight",
    "local_net.mul_blocks.1.mlp.fc1.bias": "local_net.mul_blocks.1.mlp.fc1.bias",
    "local_net.mul_blocks.1.mlp.fc2.weight": "local_net.mul_blocks.1.mlp.fc2.weight",
    "local_net.mul_blocks.1.mlp.fc2.bias": "local_net.mul_blocks.1.mlp.fc2.bias",
    "local_net.mul_blocks.2.gamma_1": "local_net.mul_blocks.2.gamma_1",
    "local_net.mul_blocks.2.gamma_2": "local_net.mul_blocks.2.gamma_2",
    "local_net.mul_blocks.2.pos_embed.weight": "local_net.mul_blocks.2.pos_embed.weight",
    "local_net.mul_blocks.2.pos_embed.bias": "local_net.mul_blocks.2.pos_embed.bias",
    "local_net.mul_blocks.2.norm1.alpha": "local_net.mul_blocks.2.norm1.alpha",
    "local_net.mul_blocks.2.norm1.beta": "local_net.mul_blocks.2.norm1.beta",
    "local_net.mul_blocks.2.norm1.color": "local_net.mul_blocks.2.norm1.color",
    "local_net.mul_blocks.2.conv1.weight": "local_net.mul_blocks.2.conv1.weight",
    "local_net.mul_blocks.2.conv1.bias": "local_net.mul_blocks.2.conv1.bias",
    "local_net.mul_blocks.2.conv2.weight": "local_net.mul_blocks.2.conv2.weight",
    "local_net.mul_blocks.2.conv2.bias": "local_net.mul_blocks.2.conv2.bias",
    "local_net.mul_blocks.2.attn.weight": "local_net.mul_blocks.2.attn.weight",
    "local_net.mul_blocks.2.attn.bias": "local_net.mul_blocks.2.attn.bias",
    "local_net.mul_blocks.2.norm2.alpha": "local_net.mul_blocks.2.norm2.alpha",
    "local_net.mul_blocks.2.norm2.beta": "local_net.mul_blocks.2.norm2.beta",
    "local_net.mul_blocks.2.norm2.color": "local_net.mul_blocks.2.norm2.color",
    "local_net.mul_blocks.2.mlp.fc1.weight": "local_net.mul_blocks.2.mlp.fc1.weight",
    "local_net.mul_blocks.2.mlp.fc1.bias": "local_net.mul_blocks.2.mlp.fc1.bias",
    "local_net.mul_blocks.2.mlp.fc2.weight": "local_net.mul_blocks.2.mlp.fc2.weight",
    "local_net.mul_blocks.2.mlp.fc2.bias": "local_net.mul_blocks.2.mlp.fc2.bias",
    "local_net.add_blocks.0.gamma_1": "local_net.add_blocks.0.gamma_1",
    "local_net.add_blocks.0.gamma_2": "local_net.add_blocks.0.gamma_2",
    "local_net.add_blocks.0.pos_embed.weight": "local_net.add_blocks.0.pos_embed.weight",
    "local_net.add_blocks.0.pos_embed.bias": "local_net.add_blocks.0.pos_embed.bias",
    "local_net.add_blocks.0.norm1.alpha": "local_net.add_blocks.0.norm1.alpha",
    "local_net.add_blocks.0.norm1.beta": "local_net.add_blocks.0.norm1.beta",
    "local_net.add_blocks.0.norm1.color": "local_net.add_blocks.0.norm1.color",
    "local_net.add_blocks.0.conv1.weight": "local_net.add_blocks.0.conv1.weight",
    "local_net.add_blocks.0.conv1.bias": "local_net.add_blocks.0.conv1.bias",
    "local_net.add_blocks.0.conv2.weight": "local_net.add_blocks.0.conv2.weight",
    "local_net.add_blocks.0.conv2.bias": "local_net.add_blocks.0.conv2.bias",
    "local_net.add_blocks.0.attn.weight": "local_net.add_blocks.0.attn.weight",
    "local_net.add_blocks.0.attn.bias": "local_net.add_blocks.0.attn.bias",
    "local_net.add_blocks.0.norm2.alpha": "local_net.add_blocks.0.norm2.alpha",
    "local_net.add_blocks.0.norm2.beta": "local_net.add_blocks.0.norm2.beta",
    "local_net.add_blocks.0.norm2.color": "local_net.add_blocks.0.norm2.color",
    "local_net.add_blocks.0.mlp.fc1.weight": "local_net.add_blocks.0.mlp.fc1.weight",
    "local_net.add_blocks.0.mlp.fc1.bias": "local_net.add_blocks.0.mlp.fc1.bias",
    "local_net.add_blocks.0.mlp.fc2.weight": "local_net.add_blocks.0.mlp.fc2.weight",
    "local_net.add_blocks.0.mlp.fc2.bias": "local_net.add_blocks.0.mlp.fc2.bias",
    "local_net.add_blocks.1.gamma_1": "local_net.add_blocks.1.gamma_1",
    "local_net.add_blocks.1.gamma_2": "local_net.add_blocks.1.gamma_2",
    "local_net.add_blocks.1.pos_embed.weight": "local_net.add_blocks.1.pos_embed.weight",
    "local_net.add_blocks.1.pos_embed.bias": "local_net.add_blocks.1.pos_embed.bias",
    "local_net.add_blocks.1.norm1.alpha": "local_net.add_blocks.1.norm1.alpha",
    "local_net.add_blocks.1.norm1.beta": "local_net.add_blocks.1.norm1.beta",
    "local_net.add_blocks.1.norm1.color": "local_net.add_blocks.1.norm1.color",
    "local_net.add_blocks.1.conv1.weight": "local_net.add_blocks.1.conv1.weight",
    "local_net.add_blocks.1.conv1.bias": "local_net.add_blocks.1.conv1.bias",
    "local_net.add_blocks.1.conv2.weight": "local_net.add_blocks.1.conv2.weight",
    "local_net.add_blocks.1.conv2.bias": "local_net.add_blocks.1.conv2.bias",
    "local_net.add_blocks.1.attn.weight": "local_net.add_blocks.1.attn.weight",
    "local_net.add_blocks.1.attn.bias": "local_net.add_blocks.1.attn.bias",
    "local_net.add_blocks.1.norm2.alpha": "local_net.add_blocks.1.norm2.alpha",
    "local_net.add_blocks.1.norm2.beta": "local_net.add_blocks.1.norm2.beta",
    "local_net.add_blocks.1.norm2.color": "local_net.add_blocks.1.norm2.color",
    "local_net.add_blocks.1.mlp.fc1.weight": "local_net.add_blocks.1.mlp.fc1.weight",
    "local_net.add_blocks.1.mlp.fc1.bias": "local_net.add_blocks.1.mlp.fc1.bias",
    "local_net.add_blocks.1.mlp.fc2.weight": "local_net.add_blocks.1.mlp.fc2.weight",
    "local_net.add_blocks.1.mlp.fc2.bias": "local_net.add_blocks.1.mlp.fc2.bias",
    "local_net.add_blocks.2.gamma_1": "local_net.add_blocks.2.gamma_1",
    "local_net.add_blocks.2.gamma_2": "local_net.add_blocks.2.gamma_2",
    "local_net.add_blocks.2.pos_embed.weight": "local_net.add_blocks.2.pos_embed.weight",
    "local_net.add_blocks.2.pos_embed.bias": "local_net.add_blocks.2.pos_embed.bias",
    "local_net.add_blocks.2.norm1.alpha": "local_net.add_blocks.2.norm1.alpha",
    "local_net.add_blocks.2.norm1.beta": "local_net.add_blocks.2.norm1.beta",
    "local_net.add_blocks.2.norm1.color": "local_net.add_blocks.2.norm1.color",
    "local_net.add_blocks.2.conv1.weight": "local_net.add_blocks.2.conv1.weight",
    "local_net.add_blocks.2.conv1.bias": "local_net.add_blocks.2.conv1.bias",
    "local_net.add_blocks.2.conv2.weight": "local_net.add_blocks.2.conv2.weight",
    "local_net.add_blocks.2.conv2.bias": "local_net.add_blocks.2.conv2.bias",
    "local_net.add_blocks.2.attn.weight": "local_net.add_blocks.2.attn.weight",
    "local_net.add_blocks.2.attn.bias": "local_net.add_blocks.2.attn.bias",
    "local_net.add_blocks.2.norm2.alpha": "local_net.add_blocks.2.norm2.alpha",
    "local_net.add_blocks.2.norm2.beta": "local_net.add_blocks.2.norm2.beta",
    "local_net.add_blocks.2.norm2.color": "local_net.add_blocks.2.norm2.color",
    "local_net.add_blocks.2.mlp.fc1.weight": "local_net.add_blocks.2.mlp.fc1.weight",
    "local_net.add_blocks.2.mlp.fc1.bias": "local_net.add_blocks.2.mlp.fc1.bias",
    "local_net.add_blocks.2.mlp.fc2.weight": "local_net.add_blocks.2.mlp.fc2.weight",
    "local_net.add_blocks.2.mlp.fc2.bias": "local_net.add_blocks.2.mlp.fc2.bias",
    "local_net.mul_end.0.weight": "local_net.mul_end.0.weight",
    "local_net.mul_end.0.bias": "local_net.mul_end.0.bias",
    "local_net.add_end.0.weight": "local_net.add_end.0.weight",
    "local_net.add_end.0.bias": "local_net.add_end.0.bias",
    "global_net.gamma_base": "global_net.gamma_base",
    "global_net.color_base": "global_net.color_base",
    "global_net.conv_large.proj.0.weight": "global_net.conv_large.proj.0.weight",
    "global_net.conv_large.proj.0.bias": "global_net.conv_large.proj.0.bias",
    "global_net.conv_large.proj.1.weight": "global_net.conv_large.proj.1.weight",
    "global_net.conv_large.proj.1.bias": "global_net.conv_large.proj.1.bias",
    "global_net.conv_large.proj.1._mean": "global_net.conv_large.proj.1.running_mean",
    "global_net.conv_large.proj.1._variance": "global_net.conv_large.proj.1.running_var",
    "global_net.conv_large.proj.3.weight": "global_net.conv_large.proj.3.weight",
    "global_net.conv_large.proj.3.bias": "global_net.conv_large.proj.3.bias",
    "global_net.conv_large.proj.4.weight": "global_net.conv_large.proj.4.weight",
    "global_net.conv_large.proj.4.bias": "global_net.conv_large.proj.4.bias",
    "global_net.conv_large.proj.4._mean": "global_net.conv_large.proj.4.running_mean",
    "global_net.conv_large.proj.4._variance": "global_net.conv_large.proj.4.running_var",
    "global_net.generator.pos_embed.weight": "global_net.generator.pos_embed.weight",
    "global_net.generator.pos_embed.bias": "global_net.generator.pos_embed.bias",
    "global_net.generator.norm1.weight": "global_net.generator.norm1.weight",
    "global_net.generator.norm1.bias": "global_net.generator.norm1.bias",
    "global_net.generator.attn.q": "global_net.generator.attn.q",
    "global_net.generator.attn.k.weight": "global_net.generator.attn.k.weight",
    "global_net.generator.attn.v.weight": "global_net.generator.attn.v.weight",
    "global_net.generator.attn.proj.weight": "global_net.generator.attn.proj.weight",
    "global_net.generator.attn.proj.bias": "global_net.generator.attn.proj.bias",
    "global_net.generator.norm2.weight": "global_net.generator.norm2.weight",
    "global_net.generator.norm2.bias": "global_net.generator.norm2.bias",
    "global_net.generator.mlp.fc1.bias": "global_net.generator.mlp.fc1.bias",
    "global_net.generator.mlp.fc2.bias": "global_net.generator.mlp.fc2.bias",
    "global_net.gamma_linear.bias": "global_net.gamma_linear.bias",
    "global_net.color_linear.bias": "global_net.color_linear.bias",
}



# 遍历 PyTorch 的 state_dict 并转换为 PaddlePaddle 格式
for torch_key in torch_state_dict:
    paddle_key = torch_key  # 默认情况下使用相同的键名
    for k in keys_dict:
        if k in paddle_key:
            # 替换 keys_dict 中定义的键名
            paddle_key = paddle_key.replace(k, keys_dict[k])
        if paddle_key == 'global_net.conv_large.proj.0.weight':
        # PyTorch: [out_channels, in_channels, height, width] -> PaddlePaddle: 相同
            if torch_state_dict[torch_key].shape == (3, 3, 32, 3):
                # 检测到不匹配的形状，进行转置：PyTorch shape (3, 3, 32, 3) -> PaddlePaddle shape (32, 3, 3, 3)
                paddle_state_dict[paddle_key] = paddle.to_tensor(torch_state_dict[torch_key].cpu().numpy().transpose(2, 3, 0, 1))
            else:
                paddle_state_dict[paddle_key] = paddle.to_tensor(torch_state_dict[torch_key].cpu().numpy())
    
        elif paddle_key == 'global_net.conv_large.proj.3.weight':
            # PyTorch: [out_channels, in_channels, height, width] -> PaddlePaddle: 相同
            if torch_state_dict[torch_key].shape == (3, 3, 64, 32):
                # 检测到不匹配的形状，进行转置：PyTorch shape (3, 3, 64, 32) -> PaddlePaddle shape (64, 32, 3, 3)
                paddle_state_dict[paddle_key] = paddle.to_tensor(torch_state_dict[torch_key].cpu().numpy().transpose(2, 3, 0, 1))
            else:
                paddle_state_dict[paddle_key] = paddle.to_tensor(torch_state_dict[torch_key].cpu().numpy())
    
        elif paddle_key == 'global_net.generator.mlp.fc1.weight':
            # 全连接层的权重需要转置：PyTorch shape [256, 64] -> PaddlePaddle shape [64, 256]
            paddle_state_dict[paddle_key] = paddle.to_tensor(torch_state_dict[torch_key].cpu().numpy().transpose())
    
        elif paddle_key == 'global_net.generator.mlp.fc2.weight':
            # 全连接层的权重需要转置：PyTorch shape [64, 256] -> PaddlePaddle shape [256, 64]
            paddle_state_dict[paddle_key] = paddle.to_tensor(torch_state_dict[torch_key].cpu().numpy().transpose())
    
        else:
            # 其余情况：卷积层等权重，按需转置
            if ('linear' in paddle_key) or ('proj' in paddle_key) or ('vocab' in paddle_key and 'weight' in paddle_key) or \
               ("dense.weight" in paddle_key) or ('transform.weight' in paddle_key) or ('seq_relationship.weight' in paddle_key):
                # 对权重进行转置
                paddle_state_dict[paddle_key] = paddle.to_tensor(torch_state_dict[torch_key].cpu().numpy().transpose())
            else:
                # 不需要转置的权重
                paddle_state_dict[paddle_key] = paddle.to_tensor(torch_state_dict[torch_key].cpu().numpy())
    

    # 打印转换前后的权重信息
    print("torch: ", torch_key, "\t", torch_state_dict[torch_key].shape)
    print("paddle: ", paddle_key, "\t", paddle_state_dict[paddle_key].shape, "\n")

# 保存转换后的 PaddlePaddle 模型权重
paddle.save(paddle_state_dict, 'transform_paddle.pdparams')
```
